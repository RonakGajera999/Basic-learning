```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
btc=pd.read_csv("C:\\python csv file\\bitcoin_price.csv")
ether=pd.read_csv("C:\\python csv file\\ethereum_price.csv")
lite_coin=pd.read_csv("C:\\python csv file\\litecoin_price.csv")
monero=pd.read_csv("C:\\python csv file\\monero_price.csv")
qutam=pd.read_csv("C:\\python csv file\\qtum_price.csv")
ripple=pd.read_csv("C:\\python csv file\\ripple_price.csv")
neo=pd.read_csv("C:\\python csv file\\neo_price.csv")
```

```python
btc.columns=btc.columns.map(lambda x : str(x) + '_btc')
ether.columns=ether.columns.map(lambda x : str(x) + '_ether')
lite_coin.columns=lite_coin.columns.map(lambda x : str(x) + '_lite_coin')
monero.columns=monero.columns.map(lambda x : str(x) + '_monero')
qutam.columns=qutam.columns.map(lambda x : str(x) + '_qutam')
ripple.columns=ripple.columns.map(lambda x : str(x) + '_ripple')
neo.columns=neo.columns.map(lambda x : str(x) + '_neo')
 
btc.head()
```

```python
#merge all data wirth date
m1=pd.merge(btc,ether,how='inner',left_on='Date_btc',right_on="Date_ether")
m2=pd.merge(m1,lite_coin,how='inner',left_on='Date_btc',right_on="Date_lite_coin")
m3=pd.merge(m2,monero,how='inner',left_on='Date_btc',right_on="Date_monero")
m4=pd.merge(m3,qutam,how='inner',left_on='Date_btc',right_on="Date_qutam")
m5=pd.merge(m4,ripple,how='inner',left_on='Date_btc',right_on="Date_ripple")
crypto=pd.merge(m5,neo,how='inner',left_on='Date_btc',right_on="Date_neo")

crypto.head()
```

```python
curr=crypto[["Close_btc","Close_ether",'Close_lite_coin','Close_monero','Close_ripple','Close_neo']]
curr.head()
```

```python
sns.pairplot(curr)
plt.show()
```

```python
cor=curr.corr()
round(cor,3)
```
